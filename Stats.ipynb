**3.What is the meaning of the five-number summary in Statistics? Give proper example**

Five number summary is very useful to analyse the distribution of the data. It contains only five numbers; these are Minimum,First Quartile(Q1),Median (Q2,Third Quartile (Q3) and Maximum.
Minimum: It shows the smallest value of the dataset.
First Quartile(q1): It shows the first 25% of the data.
Median: it shows 0 -50% of the data.
Third Quartile(q3): It show the 0-75% of the data
This number can be visualised easily with a box and whiskers plot.


**4.What is correlation? Give an example with a dataset & graphical representation on jupyter Notebook**

Correlation is the relationship between two variable one affecting other variable.

There is 2 types of correlations:
1. Positive correlation : If one variable increases, the other variable also increases linearly then there is positive correlation.
2. Negative correlation : If one variable increases, the other variable decreases linearly then there is negative correlation.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    'Hours Studied': [2, 3, 4, 5, 6, 7, 8, 9, 10],
    'Test Score': [40, 50, 60, 65, 70, 75, 80, 85, 90]
}

df = pd.DataFrame(data)

# correlation matrix
correlation_matrix = df.corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
